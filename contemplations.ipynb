```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 4))
x_present = 0.0

ax.axvline(x=x_present, color='black', linestyle='--', label='Нормик')

ax.axvspan(-1, x_present, facecolor='red', alpha=0.3, label='Стыдно')
ax.axvspan(x_present, 1, facecolor='blue', alpha=0.3, label='Страшно')

ax.text(-0.5, 0.5, 'Прошлое', ha='center', va='center', fontsize=14, color='red')
ax.text(0.50, 0.5, 'Будущее', ha='center', va='center', fontsize=14, color='blue')
ax.text(x_present, 0.7, 'Настоящее', ha='center', va='center', fontsize=12, color='black')

ax.set_xlabel("Время")

ax.legend(loc='upper left')

plt.show()
```
