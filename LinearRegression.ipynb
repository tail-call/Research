# Linear Regression

Linear regression is a fundamental statistical learning algorithm that you *must know* or you ain't a real data scientist. This notebooks aims to convert you, a dear reader, from a fake data scientist to a real data scientist if you don't know this algorithm yet â€” otherwise it will try to amuse you by reminding you of the intricacies of this simple method.

```python
ponies = [{
    'name': 'Twilight Sparkle',
    'income': 4000,
    'age': 25,
}, {
    'name': 'Fluttershy',
    'income': 1200,
    'age': 30,
}, {
    'name': 'Rarity',
    'income': 3000,
    'age': 27,
}, {
    'name': 'Pinkie Pie',
    'income': 1000,
    'age': 29,
}, {
    'name': 'Rainbow Dash',
    'income': 2000,
    'age': 24,
}, {
    'name': 'Applejack',
    'age': 25,
    'income': 500,
}]

def mean(items):
    return sum(items) / len(items)

mean_income = mean([pony['income'] for pony in ponies])
mean_age = mean([pony['age'] for pony in ponies])

(mean_income, mean_age)
```

```python
slope = sum([
    (pony['age'] - mean_age) * (pony['income'] - mean_income)
    for pony in ponies
]) / sum([
    (pony['age'] - mean_age) ** 2
    for pony in ponies
])

slope
```

```python
y_intercept = mean_income - slope * mean_age

y_intercept
```

```python
import matplotlib.pyplot as plt
import numpy as np

ages = [pony['age'] for pony in ponies]
incomes = [pony['income'] for pony in ponies]
names = [pony['name'] for pony in ponies]

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size for better viewing
plt.scatter(ages, incomes, label='Ponies', color='blue', s=50)  # s controls dot size

# Generate x-values for the regression line
x_min = min(ages) -1
x_max = max(ages) +1
x_vals = np.linspace(x_min, x_max, 100)

# Calculate corresponding y-values using the regression line equation
y_vals = slope * x_vals + y_intercept

# Plot the regression line
plt.plot(x_vals, y_vals, color='red', label='Regression Line')

# Plot the mean point
plt.scatter(mean_age, mean_income, color='green', marker='*', s=150, label='Mean (Age, Income)')

for i, name in enumerate(names):
    plt.annotate(name, (ages[i], incomes[i]), textcoords="offset points", xytext=(5,5), ha='left', fontsize=8)

plt.annotate('Mean', (mean_age, mean_income), textcoords="offset points", xytext=(5,5), ha='left', fontsize=8)


# Add labels and title
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Pony Age vs. Income with Regression Line')

plt.grid(True, linestyle='--', alpha=0.5) # Add a subtle grid
plt.legend()
plt.tight_layout() # Adjust layout to make everything fit nicely
# Show the plot
plt.show()
```

As evident from our research, ponies tend to get poorer as they age.
