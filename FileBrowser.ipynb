```python
import os

import pandas as pd

from IPython.display import display

def ls(
    path: str,
    *,
    only_dirs: bool = False
):
    os.chdir(path)

    def format_size(size_in_bytes):
        units = ['B', 'KB', 'MB', 'GB', 'TB', 'PB']
        size = size_in_bytes
        unit_index = 0

        while size >= 1024 and unit_index < len(units) - 1:
            size /= 1024
            unit_index += 1

        return f"{size:.2f} {units[unit_index]}"

    df = pd.DataFrame()
    df['Name'] = pd.Series(os.listdir())
    df['IsFile'] = df['Name'].map(lambda x: os.path.isfile(x))
    df['Kind'] = df['Name'].map(lambda x: os.path.splitext(x)[1][1:] or '/')
    df['SizeHR'] = df['Name'].map(lambda x: format_size(os.path.getsize(x)))
    df['Size'] = df['Name'].map(lambda x: os.path.getsize(x))
    df = df.sort_values(by='Size', ascending=False)

    if only_dirs:
        df = df[df['IsFile'] == False]

    print(f"Path: {path} | Total: {format_size(df.Size.sum())}")
    display(df)
    
ls('/Users/scales/Desktop/Hoard/PDF')
```
