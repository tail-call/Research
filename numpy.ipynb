## Operations

Element-wise product:

```python
c = np.array([[1, 2],
              [3, 4]])
d = np.array([[5, 6],
              [7, 8]])

c * d
```

Proper matrix multiplication:

```python
e = np.array([[1, 2],
              [3, 4],
              [5, 6]])
f = np.array([[7, 8],
              [9,10]])

e @ f

# This also works:
e.dot(f)

# This won't work:
# >>> f @ e
# 
# For matrix multiplication to work, left hand width must be equal to
# right hand height.
# >>> (e.shape, f.shape)
# ((3, 2), (2, 2)) # OK: 2 == 2
#
# >>> (f.shape, e.shape)
# ((2, 2), (3, 2)) # Not OK: 2 != 3
```

Also we can add/subtract scalars.

```python
(c * d) - 14
```

Now look at this:

```python
((c * d) - 14) > 0
```

```python
g = e.dot(f)
g
```
